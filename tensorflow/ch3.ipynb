# Minimizing Cost

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
x_train = [1, 2, 3, 4]
y_train = [0, -1, -2, -3]

tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1))

sgd = tf.keras.optimizers.SGD(lr=0.1)
tf.model.compile(loss='mse', optimizer=sgd)

# model을 요약해서 shape 등의 정보를 봄
tf.model.summary()

# fit()은 model을 트레이닝 시키고 그 과정을 반환
history = tf.model.fit(x_train, y_train, epochs=100)

y_predict = tf.model.predict(np.array([5,4]))
print(y_predict)

# 학습 표시 & loss values 시각화
plt.plot(history.history['loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD, Adam
```

```python
# 데이터 생성
# 2x(1) - 3x(2) + 2x(3) = 정답
# 입력
x_train = np.array([[1, 2, 0], [5, 4, 3], [1, 2, -1], [3, 1, 0], [2, 4, 2],
                    [4, 1, 2], [-1, 3, 2], [4, 3, 3], [0, 2, 6], [2, 2, 1],
                    [1, -2, -2], [0, 1, 3], [1, 1, 3], [0, 1, 4], [2, 3, 3]])
# 정답
y_train = np.array([-4, 4, -6, 3, -4,
                    9, -7, 5, 6, 0,
                    4, 3, 5, 5, 1])

print('x_train.shape = ', x_train.shape, ', y_train.shape = ', y_train.shape)
```

```python
# 모델 구축
model = Sequential()
model.add(Dense(1, input_shape=(3, ), activation='linear'))
```

```python
# 모델 컴파일
# SGD -> 학습 알고리즘, learning_rate -> 학습률, loss -> 손실함수
model.compile(optimizer=SGD(learning_rate=1e-2), loss='mse')
model.summary()
```

```python
# 모델 학습
hist = model.fit(x_train, y_train, epochs=1000)
```

```python
# 모델 평가 및 예측
test_data = [[5, 5, 0], [2, 3, 1], [-1, 0, -1], [10, 5, 2], [4, -1, -2]]
ret_val = [ 2*data[0] -3*data[1] + 2*data[2] for data in test_data]
prediction_val = model.predict(np.array(test_data))

print(prediction_val) # 예측값
print('=======================')
print(ret_val) # 정답
```

```python
print(model.input) # 모델 입력
print(model.output) # 모델 출력
print(model.weights) # 모델 가중치

# 'dense/kernel:0' shape(3, 1) -> 가중치
# 'dense/bias:0' shpqe(1, ) -> 바이어스
```

```python
plt.title('Loss Trend')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid()

# 측정지표 metrics=['loss'] 항목은 기본으로 지정되어 hist.history['loss']로 저장됨
plt.plot(hist.history['loss'], label='train loss')
plt.legend(loc='best')

# 손실함수가 지속적으로 감소한다는 것을 확인할 수 있음
plt.show()
```
