```python
import numpy as np
import tensorflow as tf
```

# H(x) = Wx + b

```python
# X and Y data
x_train = [1, 2, 3, 4]
y_train = [0, -1, -2 ,-3]

# keras는 기본적으로 다차원 레이를 기본으로 하고 있기 때문에 Sequential을 생성
# keras의 다차원 계층 모델인 Sequential 레이어를 만듬
tf.model = tf.keras.Sequential()

# 입력(input_dim) : 1차원, 출력(units) : 1차원, -Dense : 레이어의 종류
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1))

# optimizer : loss function을 통해 구한 차이를 사용해 기울기를 구하고 네트워크의 cost(W,b)를 학습에 어떻게 반영할 것인지 결정하는 방법
# optimizer - Stochastic gradient descent(확률적 경사 하강법) : 경사 하강법과 다르게 한번 학습할 때 모든 데이터에 대해 가중치를 조절하는 것이 아닌 랜덤하게 추출한 일부 데이터에 대해 가중치 조절, 결과적으로 속도는 개선 되었으나 최적해의 정확도는 낮음
# SGD : standard gradient descendent, lr : learning rate
sgd = tf.keras.optimizers.SGD(lr=0.1)

# cost(loss) function
# loss를 mean_squared_error 방식을 사용한다는 의미로 'mse'라고 써도 인식
# mse : 1/m * sig(y'-y)^2
tf.model.compile(loss='mse', optimizer=sgd)

# fit the line
# 버전 1과 다르게 session을 만들어 run할 필요 없음
tf.model.fit(x_train, y_train, epochs=200)

# predict()는 예측값을 반환
y_predict = tf.model.predict(np.array([5,4]))
print(y_predict)
```
